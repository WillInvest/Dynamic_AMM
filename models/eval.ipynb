```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

def eval_plot_model(path, window=50):
    evaluations = np.load(path)
    timesteps = evaluations['timesteps']
    results = evaluations['results']
    
    # Compute the mean evaluation rewards over time
    mean_results = results.mean(axis=1)
    
    # Plot the mean evaluation rewards over time
    plt.plot(timesteps, mean_results, label='Mean Reward Over Time')

    # Compute the rolling mean (moving average)
    rolling_mean = pd.Series(mean_results).rolling(window=window).mean()
    
    # Plot the rolling mean
    plt.plot(timesteps, rolling_mean, color='orange', label=f'Rolling Mean (window={window})')
    
    plt.xlabel('Timesteps')
    plt.ylabel('Mean Evaluation Reward')
    plt.title('Model Performance Over Time')
    plt.legend()
    plt.show()

    # Sort the timesteps by the top 10 evaluation scores
    top_10_indices = np.argsort(mean_results)[-10:][::-1]  # Get indices of top 10 evaluations, sorted in descending order
    top_10_timesteps = timesteps[top_10_indices]
    top_10_results = mean_results[top_10_indices]

    top_10_df = pd.DataFrame({
        'Timestep': top_10_timesteps,
        'Mean Evaluation Reward': top_10_results
    })

    print(top_10_df)
```

```python
path = '/home/shiftpub/AMM-Python/models/dummy_maker_model/evaluations.npz'

eval_plot_model(path)
```

```python
for mc in np.arange(0.05, 1.05, 0.05):
    mc = np.round(mc, 2)

    path = f'/home/shiftpub/AMM-Python/models/trader_model/market_competition_level_{mc:.2f}/logs/evaluations.npz'

    eval_plot_model(path)
```
